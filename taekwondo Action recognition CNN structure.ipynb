```python
import os, re, glob, time, cv2
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import shutil
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from numpy import argmax
from tensorflow.python.keras import backend as K
from tensorflow.python.keras.models import Sequential, Model, load_model
from tensorflow.python.keras import layers
from tensorflow.python.keras.optimizers import SGD, Adam
from tensorflow.python.keras import regularizers
import time
import pandas as pd
base_dir = './excel/' 
print("success")
```

```python
model = keras.models.Sequential([
    
    keras.layers.Conv2D(16, kernel_size=(5, 5), activation='relu', input_shape=(200, 120, 3)),
    keras.layers.MaxPooling2D(pool_size=(2, 2)),
    keras.layers.Conv2D(32, kernel_size=(3, 3), activation='relu'),
    keras.layers.MaxPooling2D(pool_size=(2, 2)),
    keras.layers.Conv2D(64, kernel_size=(3, 3), activation='relu'),
    keras.layers.MaxPooling2D(pool_size=(2, 2)),
    keras.layers.Conv2D(128, kernel_size=(3, 3), activation='relu'),
    keras.layers.MaxPooling2D(pool_size=(2, 2)),
    keras.layers.Flatten(),
    
    keras.layers.Dense(128, activation='relu'),
    # keras.layers.BatchNormalization(),
    keras.layers.Dropout(0.2),      
    keras.layers.Dense(32, activation='softmax'),
])
print("success")
```

```python
groups_folder_pos_path = './pos/'               # './50/RGB_f/'  
groups_folder_rot_path = './rot/'               # './50/Depth_f/' 
categories = ["1L", "1R", "2L", "2R", "3L", "3R", "4L", "4R","5L", "5R", "6L", "6R", "7L", "7R", "8L", "8R","9L", "9R", "10L", "10R", "11L", "11R", "12L", "12R","13L", "13R", "14L", "14R", "15L", "15R", "16L", "16R"]
#categories = ["1L", "1R", "2L", "2R", "3L", "3R", "4L", "4R","5L", "5R", "6L", "6R", "7L", "7R", "8L", "8R"]
#categories = ["1L", "2L",  "3L", "4L", "5L", "6L",  "7L",  "8L"]
num_classes = len(categories)
len(categories)
```

```python
image_w = 120
image_h = 200
  
X = []
Y = []
test = 0
for idex, categorie in enumerate(categories):
    label = [0 for i in range(num_classes)]
    label[idex] = 1
    image_dir_pos = groups_folder_pos_path + categorie + '/'
    image_dir_rot = groups_folder_rot_path + categorie + '/'


    for top, dir, f in os.walk(image_dir_pos):
        
        for filename in f:
            try:
                pos = cv2.imread(image_dir_pos +filename)
                print(pos.shape)
                pos = cv2.resize(pos, dsize=(image_w, image_h))
                img = pos
                img =  img/256
                img = np.float16(img)
                X.append(img)
                Y.append(label)
            except:
                print("FAIL PLEASE")
                print(groups_folder_pos_path + categorie + filename)

X = np.array(X)
Y = np.array(Y)
print(num_classes)
print(X.shape)
print(Y.shape)
print("finish")
```

```python
# x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.3 ,shuffle=True)
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.3 ,shuffle=True, stratify = Y )


print(x_train.shape)
print(Y.shape)
print(y_test.shape)
```

```python
skf = KFold(n_splits=5, shuffle=True)
accuracy=[]
for train, validation in skf.split(x_train, y_train):
    print(train.shape)
    print(x_train[train].shape)
    model.fit(x_train[train], y_train[train], epochs=20, batch_size = 16) #epoch = 20  batch_size = 16
    print("test")
    k_accuracy = '%.4f' % (model.evaluate(x_train[validation], y_train[validation])[1])
    accuracy.append(k_accuracy)

print('\nK-fold cross validation Accuracy: {}'.format(accuracy))
```

```python
start = time.time()
test_loss, test_acc = model.evaluate(x_test,  y_test, verbose=1)
predict = model.predict_classes(x_test)
print("TIME is ", time.time()-start)
print('\n :', test_acc)
print(len(predict))
```
